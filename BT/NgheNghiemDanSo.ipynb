```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv("NgeNghiepDanSo.csv", encoding='utf-8')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=df, x="Ngành nghề", y="Tuổi")

plt.title("Trung bình độ tuổi theo ngành nghề")
plt.xlabel("Ngành nghề")
plt.ylabel("Độ tuổi trung bình")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Tạo dữ liệu cho pie chart
occupation_counts = df['Ngành nghề'].value_counts()

# Vẽ pie chart
fig = px.pie(values=occupation_counts.values, names=occupation_counts.index, title="Tỷ lệ phần trăm các ngành nghề")
fig.show()
```

```python
# Biểu đồ tỷ lệ giới tính
gender_counts = df['Giới tính'].value_counts()

fig = px.pie(values=gender_counts.values, names=gender_counts.index, title="Tỷ lệ giới tính")
fig.show()
```

```python
# Biểu đồ tỷ lệ tình trạng hôn nhân
marital_counts = df['Tình trạng hôn nhân'].value_counts()

fig = px.pie(values=marital_counts.values, names=marital_counts.index, title="Tỷ lệ tình trạng hôn nhân")
fig.show()
```
